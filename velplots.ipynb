```python
def readintotal(csv):
    try:
        df = pd.read_csv(csv, names = ['t','ke','gpe'], delim_whitespace=True)
        df = df.iloc[::1, :]
        df = df.iloc[::1, :]
    except:
        df = pd.DataFrame(columns=['t','ke','gpe'])
    return df
```
